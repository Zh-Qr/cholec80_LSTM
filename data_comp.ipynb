## 状态噪声比较

```python
dir_cleantool = '..\\cholec80\\annotations'
dir_noisytool = 'LSTM_SBS_state'

video_namepath = 'splite_noise\\test_videos.txt'
with open(video_namepath, 'r', encoding='utf-8') as file:
    videos  = [line.strip() for line in file]


# 初始化计数器
total_lines = 0  # 总行数
different_lines = 0  # 不同的行数

# 遍历每个视频
for video in videos:
    total = 0
    different = 0
    print(f'testint {video}.txt')
    clean_path = os.path.join(dir_cleantool, f'{video}.txt')
    noisy_path = os.path.join(dir_noisytool, f'{video}.txt')
    
    # 读取clean tool文件，跳过第一行
    with open(clean_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        clean_tool = [line.strip() for line in lines[1:]]  # 从第二行开始读取
    
    # 读取noisy tool文件，跳过第一行
    with open(noisy_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        noisy_tool = [line.strip() for line in lines[1:]]  # 从第二行开始读取
    
    # 逐行对比（假设两个文件行数一致，否则可以根据实际情况调整）
    for clean_line, noisy_line in zip(clean_tool, noisy_tool):      
        total += 1
        if clean_line != noisy_line:
            different += 1
            
    total_lines = total + total_lines
    different_lines = different + different_lines
    
    print(f"视频{video}噪声率是：{different / total:.2f}")
            

# 计算整体噪声率
print("total:", total_lines)
noise_rate = different_lines / total_lines if total_lines > 0 else 0
print("整体噪声率:", noise_rate)
```

```python
dir_cleantool = r'..\cholec80\annotations'
dir_noisytool = r'LSTM_SBS_state'

video_namepath = r'splite_noise\test_videos.txt'
with open(video_namepath, 'r', encoding='utf-8') as file:
    videos = [line.strip() for line in file]

# 初始化计数器
total_lines = 0  # 总行数
different_lines = 0  # 不同的行数

# 遍历每个视频
for video in videos:
    total = 0
    different = 0
    print(f'正在测试 {video}.txt')
    clean_path = os.path.join(dir_cleantool, f'{video}.txt')
    noisy_path = os.path.join(dir_noisytool, f'{video}.txt')
    
    # 读取 clean tool 文件，跳过第一行（表头）
    with open(clean_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        clean_tool = [line.strip() for line in lines[1:]]
    
    # 读取 noisy tool 文件，跳过第一行
    with open(noisy_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        noisy_tool = [line.strip() for line in lines[1:]]
    
    # 逐行对比（假设两个文件行数一致，否则根据实际情况调整）
    for clean_line, noisy_line in zip(clean_tool, noisy_tool):
        total += 1
        if clean_line != noisy_line:
            different += 1
            
    # 计算当前视频的准确率
    accuracy_video = (total - different) / total if total > 0 else 0
    total_lines += total
    different_lines += different
    
    print(f"视频 {video} 的准确率是：{accuracy_video:.2f}")

# 计算整体准确率
overall_accuracy = (total_lines - different_lines) / total_lines if total_lines > 0 else 0
print("总行数:", total_lines)
print("整体准确率:", overall_accuracy)
```

## state分布可视化

```python
PHASES = [
    "Preparation",
    "CalotTriangleDissection",
    "ClippingAndCutting",
    "GallbladderDissection",
    "GallbladderPackaging",
    "CleaningAndCoagulation",
    "GallbladderRetraction"
]

def load_annotations_to_list_step(filepath, step=25):
    """
    每隔 step 行读取一次，直接返回形状为 (1, T) 的 NumPy 数组，
    其中每个元素是数字。
    """
    numbers = []
    with open(filepath, "r", encoding="utf-8") as f:
        for line_index, line in enumerate(f):
            if line_index % step == 0:
                line = line.strip()
                if not line:
                    continue
                # 直接将每行内容转换为数字，这里假设每行均为整数数字
                try:
                    number = int(line)
                except ValueError:
                    # 若转换失败，则尝试转换为浮点数
                    number = float(line)
                numbers.append(number)
    encoded_array = np.array(numbers, dtype=np.int32)
    encoded_array_2d = np.expand_dims(encoded_array, axis=0)
    return encoded_array_2d

def plot_pic(phases_numeric):
    """
    phases_numeric: 2D numpy 数组，形状为 (1, T)；每个元素为阶段编码（1~7）
    """
    fig = plt.figure(figsize=(10, 2))
    ax = fig.add_subplot(111)
    ax.set_yticks([])
    ax.set_yticklabels([])
    im = ax.pcolormesh(phases_numeric, cmap="Set2", vmin=1, vmax=len(PHASES))
    ticks_mid = [i + 0.5 for i in range(1, len(PHASES) + 1)]
    cbar = fig.colorbar(im, ticks=ticks_mid, orientation="vertical", drawedges=True)
    cbar.ax.set_yticklabels(PHASES)
    plt.show()
```

```python
video_id = 79
if video_id<10:
    clean_filepath = f'..\\cholec80\\annotations\\video0{video_id}.txt'
    noisy_filepath = f'LSTM_SBS_state\\video0{video_id}.txt'
else:
    clean_filepath = f'..\\cholec80\\annotations\\video{video_id}.txt'
    noisy_filepath = f'LSTM_SBS_state\\video{video_id}.txt'
clean_ann = load_annotations_to_list_step(clean_filepath)
noisy_ann = load_annotations_to_list_step(noisy_filepath)  # 修正变量名

plot_pic(clean_ann)
plot_pic(noisy_ann)

# 统计噪声帧率
min_length = min(len(clean_ann[0]), len(noisy_ann[0]))  # 注意：数组为 (1, T)，需取 [0] 索引
noise_frames = 0
for i in range(min_length):
    # 比较对应位置的元素
    if clean_ann[0][i] != noisy_ann[0][i]:
        noise_frames += 1

total_frames = min_length
print(f'噪声率 {noise_frames/total_frames:.2f}')
```

## 卷积神经网络关注点对比图

```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image
import matplotlib.pyplot as plt

# -------------------------
# 模型与预处理设置（与已有代码一致）
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_classes = 7
model = models.resnet50(pretrained=True)
in_features = model.fc.in_features
model.fc = nn.Linear(in_features, num_classes)
model = model.to(device)

checkpoint_path = 'weight\\tool_detection.pth'
# checkpoint_path = 'weight\\generate_loss.pth'

state_dict = torch.load(checkpoint_path, map_location=device)
model.load_state_dict(state_dict, strict=False)
model.eval()

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

video_id = '02' 
image_path = f'..\\cholec80\\frames\\video{video_id}'

# -------------------------
# Grad-CAM 实现
# -------------------------
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None
        self.hook_handles = []
        self._register_hooks()
        
    def _register_hooks(self):
        def forward_hook(module, input, output):
            self.activations = output.detach()
        
        def backward_hook(module, grad_in, grad_out):
            self.gradients = grad_out[0].detach()
        
        handle_forward = self.target_layer.register_forward_hook(forward_hook)
        handle_backward = self.target_layer.register_backward_hook(backward_hook)
        self.hook_handles.extend([handle_forward, handle_backward])
    
    def remove_hooks(self):
        for handle in self.hook_handles:
            handle.remove()
    
    def __call__(self, input_tensor, class_idx=None):
        # 前向传播
        output = self.model(input_tensor)
        if class_idx is None:
            class_idx = output.argmax(dim=1).item()
        # 梯度清零
        self.model.zero_grad()
        # 对预测类别做反向传播
        one_hot = torch.zeros_like(output)
        one_hot[0, class_idx] = 1
        output.backward(gradient=one_hot, retain_graph=True)
        # 计算权重（对梯度做全局平均池化）
        gradients = self.gradients  # [batch, channels, h, w]
        activations = self.activations  # [batch, channels, h, w]
        weights = gradients.mean(dim=(2, 3), keepdim=True)
        # Grad-CAM 热力图
        grad_cam_map = torch.relu((weights * activations).sum(dim=1, keepdim=True))
        grad_cam_map = torch.nn.functional.interpolate(grad_cam_map, size=input_tensor.shape[2:], mode='bilinear', align_corners=False)
        grad_cam_map = grad_cam_map - grad_cam_map.min()
        grad_cam_map = grad_cam_map / (grad_cam_map.max() + 1e-8)
        return grad_cam_map, class_idx

# 使用 ResNet50 的 layer4 作为目标层
gradcam = GradCAM(model, model.layer4)

# -------------------------
# 选择第 n 张图片，并进行预测与 Grad-CAM 可视化
# -------------------------
n = 293  # 选择第 n 张图片（例如第 10 张），注意：此处按排序后的列表索引（从 0 开始）
image_files = sorted(os.listdir(image_path))
if n >= len(image_files):
    raise ValueError("n 超过了文件夹中图片的数量")
selected_file = image_files[n]
img_full_path = os.path.join(image_path, selected_file)

# 加载图片，并进行预处理
original_image = Image.open(img_full_path).convert('RGB')
input_tensor = test_transform(original_image).unsqueeze(0).to(device)

# 利用 Grad-CAM 得到热力图及预测类别
grad_cam_map, predicted_class = gradcam(input_tensor)
print("Predicted class:", predicted_class)

# 将热力图转换为 NumPy 格式并调整到原图尺寸
heatmap = grad_cam_map.cpu().numpy()[0, 0]
# 注意：原图尺寸与 test_transform 中 Resize(224,224) 后不同，此处需要用原图尺寸进行 resize
heatmap = cv2.resize(heatmap, original_image.size)
heatmap = np.uint8(255 * heatmap)
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

# 将原图转换为 OpenCV 格式，并进行颜色空间转换（RGB -> BGR）
original_image_cv = cv2.cvtColor(np.array(original_image), cv2.COLOR_RGB2BGR)
# 叠加热力图（可以调整权重使得高亮区域更明显）
overlay = cv2.addWeighted(original_image_cv, 0.6, heatmap, 0.4, 0)

# 使用 matplotlib 显示叠加后的图像
plt.figure(figsize=(10, 10))
# 将 BGR 转换回 RGB 后显示
plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title(f'Grad-CAM Overlay (Predicted Class: {predicted_class})')
plt.savefig('picture\\Grad-CAM01.jpg')
plt.show()

# 清除 hook
gradcam.remove_hooks()
```

## log数据对比可视化

```python
def plot_log_pic(train_log, flag = False, Name=None):
    plt.figure()
    
    # 子图1：训练损失（第二列）
    plt.subplot(2, 2, 1)
    plt.plot(train_log.iloc[:, 1], "r", label='Train Loss')
    plt.title('Train Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)

    # 子图2：训练准确率（第三列）
    plt.subplot(2, 2, 2)
    plt.plot(train_log.iloc[:, 2], "g", label='Train Accuracy')
    plt.title('Train Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)

    # 子图3：测试损失（第四列）
    plt.subplot(2, 2, 3)
    plt.plot(train_log.iloc[:, 3], "b", label='Test Loss')
    plt.title('Test Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)

    # 子图4：测试准确率（第五列）
    plt.subplot(2, 2, 4)
    plt.plot(train_log.iloc[:, 4], "y", label='Test Accuracy')
    plt.title('Test Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    if flag:
        plt.savefig(Name)
        print(f"图片已经保存在{Name}")
    plt.show()


def comp_plot_log_pic(train_log1, train_log2, title, name1, name2, flag = False, Name=None):
    plt.figure()
    fig = plt.gcf()
    fig.suptitle(title, fontsize=16, fontweight='bold')
    
    # 子图1：对比训练准确率（第三列）
    plt.subplot(1, 2, 1)
    plt.plot(train_log1.iloc[:, 2], "r", label=f'{name1} Train Accuracy')
    plt.plot(train_log2.iloc[:, 2], "g", label=f'{name2} Train Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.title('Train Accuracy')
    
    # 子图2：对比测试准确率（第五列）
    plt.subplot(1, 2, 2)
    plt.plot(train_log1.iloc[:, 4], "r", label=f'{name1} Test Accuracy')
    plt.plot(train_log2.iloc[:, 4], "g", label=f'{name2} Test Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid(True)
    plt.title('Test Accuracy')
    
    plt.tight_layout()
    if flag:
        plt.savefig(Name)
        print(f"图片已经保存在{Name}")
    plt.show()
```
